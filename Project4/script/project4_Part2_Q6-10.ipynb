```r
library(jsonlite)
library(igraph)
# read the data file
json_data <- fromJSON('los_angeles_censustracts.json')

# save display name and locations as features of each id
node_num = dim(json_data[["features"]][["properties"]])[1]
longitude = c()
latitude = c()
for (i in 1:node_num) {
    l = length(json_data[["features"]][["geometry"]][["coordinates"]][[i]])
    longitude <- c(longitude,mean(json_data[["features"]][["geometry"]][["coordinates"]][[i]][1:(l/2)]))
    latitude <- c(latitude,mean(json_data[["features"]][["geometry"]][["coordinates"]][[i]][(l/2+1):l]))
}
node_info <- data.frame(id=json_data[["features"]][["properties"]][["MOVEMENT_ID"]],
                     display_name=json_data[["features"]][["properties"]][["DISPLAY_NAME"]],
                     longitude=longitude, latitude=latitude)
```

```r

g <- read_graph('uber_gcc.txt', format='ncol', directed=FALSE)

# take average mean travel time for duplicate edges
g <- simplify(g, edge.attr.comb="mean")
print(g)
```

```r
# select out GCC for future use
g.components <- clusters(g)
ix <- which.max(g.components$csize)
g.giant <- induced.subgraph(g, which(g.components$membership == ix))
g.giant

# Q6. Report the number of nodes and edges in G.
print(vcount(g.giant))
print(ecount(g.giant))
```

```r
# Q8. Determine what percentage of triangles in the graph (sets of 3 points on the map) satisfy the triangle inequality.
gg_tri = matrix(triangles(g.giant), ncol = 3)
gg_tri_sub = gg_tri[sample(nrow(gg_tri),1000,replace=TRUE),]      # randomly sample 1000 triangles

n = 0
for (i in 1:1000) {
    v1 = gg_tri_sub[i,1]
    v2 = gg_tri_sub[i,2]
    v3 = gg_tri_sub[i,3]
    e1 = g.giant[from=v1, to=v2, attr="weight"]
    e2 = g.giant[from=v1, to=v3, attr="weight"]
    e3 = g.giant[from=v2, to=v3, attr="weight"]
    # test if it satisfies the triangle inequality.
    if (e1 + e2 > e3 | e1 + e3 > e2 | e2 + e3 > e1) {
        n = n + 1
    }
}
perc = n /1000
print(perc)
```

```r
# Q9. Find an upper bound on the empirical performance of the approximate algorithm.
mst_cost = sum(E(g_mst)$weight)        # MST cost

# approximate TSP cost
mst_dfs <-dfs(g_mst,1)        # transverse the MST from vertex 1
order <- mst_dfs$order$name
tsp_perm <- c(order,order[1])
tsp_cost = 0
for (i in 1:(length(tsp_perm)-1)) {
    v <- tsp_perm[c(i, i+1)]
    if (are_adjacent(g.giant,v[1],v[2])) {
        tsp_cost = tsp_cost + g.giant[from=v[1], to=v[2], attr="weight"]
    }
    else {
        tsp_cost = tsp_cost + distances(g.giant,v=v[1],to=v[2])[1]
    }
}


rou = tsp_cost / mst_cost
print(paste(mst_cost,tsp_cost,rou))
```

```r
# change the start vertex to transverse, and see the change of TSP cost approximation
v_sample = sample(V(g.giant),50,replace=TRUE)
tsp_cost_total = c()

for (s in v_sample$name) {
    mst_dfs <-dfs(g_mst,s)        # transverse the MST from vertex 1
    order <- mst_dfs$order$name
    tsp_perm <- c(order,order[1])
    tsp_cost = 0
    for (i in 1:(length(tsp_perm)-1)) {
        v <- tsp_perm[c(i, i+1)]
        if (are_adjacent(g.giant,v[1],v[2])) {
            tsp_cost = tsp_cost + g.giant[from=v[1], to=v[2], attr="weight"]
        }
        else {
            tsp_cost = tsp_cost + distances(g.giant,v=v[1],to=v[2])[1]
        }
    }
    tsp_cost_total = c(tsp_cost_total,tsp_cost)
}
max(tsp_cost_total)
```

```r
# then we found that some location info is NA because the format in the origianl .json is different. we found that in total
# there are 28 NA, and they have 2 different formats. so we need to deal with them seperately. thus, we reconstruct the dataframe.

node_num = dim(json_data[["features"]][["properties"]])[1]
longitude = rep(NA,node_num)
latitude = rep(NA,node_num)
for (i in 1:node_num) {
    l = length(json_data[["features"]][["geometry"]][["coordinates"]][[i]])
    longitude[i] <- mean(json_data[["features"]][["geometry"]][["coordinates"]][[i]][1:(l/2)])
    latitude[i] <- mean(json_data[["features"]][["geometry"]][["coordinates"]][[i]][(l/2+1):l])
}
na_id = which(is.na(longitude))
for (i in na_id) {
    coor = json_data[["features"]][["geometry"]][["coordinates"]][[i]]
    l_total = 0
    part_lon = 0
    part_lat = 0
    if (length(coor) == 1) {
        # type 1
        for (j in 1:length(coor[[1]])) {
            l = length(coor[[1]][[j]])
            part_lon = part_lon + sum(coor[[1]][[j]][1:(l/2)])
            part_lat = part_lat + sum(coor[[1]][[j]][(l/2+1):l])
            l_total = l_total + l/2
        }
        longitude[i] = part_lon/l_total
        latitude[i] = part_lat/l_total
    } else {
        # type 2
        for (j in 1:length(coor)) {
            l = length(coor[[j]])
            part_lon = part_lon + sum(coor[[j]][1:(l/2)])
            part_lat = part_lat + sum(coor[[j]][(l/2+1):l])
            l_total = l_total + l/2
        }
        longitude[i] = part_lon/l_total
        latitude[i] = part_lat/l_total
    }
}
node_info <- data.frame(id=json_data[["features"]][["properties"]][["MOVEMENT_ID"]],
                     display_name=json_data[["features"]][["properties"]][["DISPLAY_NAME"]],
                     longitude=longitude, latitude=latitude)
```
