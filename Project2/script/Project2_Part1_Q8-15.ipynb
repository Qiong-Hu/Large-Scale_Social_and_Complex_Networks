```r
# Embeddedness function
embed <- function(personal_graph, Vnoncore) {
    return(degree(personal_graph, Vnoncore)-1)
}
```

```r
# Dispersion function
disp <- function(personal_graph, Vcore, Vnoncore) {
    # personal_graph include Vcore and Vnoncore
    # Vcore is the node ID in whole graph
    # Vnoncore is the node index in the personalized graph
    # Vcore = Vnoncore is possible and acceptable
    # embeddedness of Vnoncore should be non-zero
    
    dis <- 0
    if (embed(personal_graph, Vnoncore) < 2) {
        return(NA)
    }
    else {
        mutual_friend_list <- neighbors(personal_graph, Vnoncore)[-which(neighbors(personal_graph, Vnoncore)$name==Vcore)]
        mutual_friend_list <- as.numeric(mutual_friend_list$name)
        sub_graph <- induced_subgraph(personal_graph, V(personal_graph)[-c(which(V(personal_graph)$name==Vcore),Vnoncore)])
        
        mutual_friend_num <- length(mutual_friend_list)
        for (i in 1:(mutual_friend_num-1)) {
            for (j in (i+1):mutual_friend_num) {
                if (as.numeric(distances(sub_graph,which(V(sub_graph)$name==mutual_friend_list[i]),which(V(sub_graph)$name==mutual_friend_list[j]))) > 2) {
                    dis <- dis + 1
                }
            }
        }
        return (dis)
    }
}
```

```r
plot_highlight <- function(graph, community, V_core, V_highlight, title) {
    # V_core is the core node's ID in fb_graph
    # V_highlight is the highlighted node's index in personalized network

    V(graph)$color <- "gray"
    V(graph)[which(V(graph)$name==V_core)]$color <- "orange"
    V(graph)[V_highlight]$color <- "red"

    V(graph)$size <- 3
    V(graph)[which(V(graph)$name==V_core)]$size <- 5
    V(graph)[V_highlight]$size <- 5
    
    V(graph)$label <- NA
    for (each_highlight in V_highlight) {
        V(graph)[each_highlight]$label <- V(graph)[each_highlight]$name
    }

    E(graph)$color <- "gray"
    E(graph)$width <- 1
    for (each_highlight in V_highlight) {
        E(graph)[match(incident(graph,each_highlight),E(graph))]$color <- "red"
        E(graph)[match(incident(graph,each_highlight),E(graph))]$width <- 3
    }

    plot(community, graph, 
#          vertex.label = NA, 
         vertex.color = V(graph)$color, 
         vertex.size = V(graph)$size, 
         edge.color = E(graph)$color, 
         edge.width = E(graph)$width, 
         main = title)
}
```
